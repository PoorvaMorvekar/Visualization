# Graph: the Curves



- Read 'unrate.csv' in a dataframe and assign this result to the variable unrate.
- Use the pandas.to_datetime() function to convert the 'DATE' column to datetime values.
- Display the first 12 unrate values corresponding to a whole year.


```python
import pandas as pd
unrate=pd.read_csv('unrate.csv')
unrate['DATE']=pd.to_datetime(unrate['DATE'])
unrate.head(12)
```

## Adding data to our charts


- Generate a line graph to visualize the unemployment rates for 1948:
>- Take for x the first 12 values in the column 'DATE'.
>- Take for y the first 12 values of the column 'UNRATE'.
- Display the graph.


```python
import matplotlib.pyplot as plt
plt.plot(unrate.DATE.head(12),unrate.UNRATE.head(12))
plt.show()
```

## Create a rotation on the x-axis data


- Generate the same line graph to visualize the unemployment rate in 1948:
>- take for x the first 12 values in the column 'DATE'.
>- take for y the first 12 values of the column 'UNRATE'.
- Use the pyplot.xticks() function to create a 90-degree roration.
- Display the graph.


```python
plt.plot(unrate.DATE.head(12),unrate.UNRATE.head(12))
plt.xticks(rotation=90)
plt.show()
```

## Add axis headings and a title to our chart ##


- Generate the same graph as before by adding the x and y axis headings and the title.
- Define the name of the x-axis as "Month".
- Define the name of the y-axis as "Unemployment Rate (%)".
- Define the title as "Evolution of the unemployment rate in 1948 in the USA".
- Display the graph.


```python
plt.title('Evolution of the unemployment rate in 1948 in the USA')
plt.plot(unrate.DATE.head(12),unrate.UNRATE.head(12))
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

## Adding data to graphs


Create 2 linear curves in a 2 row 1 column grid:
- In the top sub-graph, plot the 1948 data.
>- For the x-axis, take the first 12 values of the column 'DATE'.
>- Take the first 12 values of the 'UNRATE' column for the y axis.
- In the bottom sub-graph, plot the 1949 data.
>- For the x-axis, take the index values 12 to 24 from the 'DATE' column (the next 12 values in the dataframe).
>- Take the index values 12 to 24 from the 'UNRATE' column for the y axis.

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2)
axs[0].plot(unrate.DATE.head(12),unrate.UNRATE.head(12))
axs[1].plot(unrate[12:24].DATE,unrate[12:24].UNRATE)
plt.xticks(rotation=90)
plt.show()
```

## Improve the display

## Superimposing curves


- Set the display area to 6 inches wide and 3 inches high.
- Generate 2 curves in a single graph using the 'MONTH' column for the x-axis instead of the 'DATE' column:
>- One curve with 1948 values in 'red' color.
>- A curve with the 1949 values of color 'blue'.
- Use plt.show() to display the curves.


```python
fig = plt.figure(figsize=(6,3))
for i in range(1950-1948):
    plt.plot(unrate[(i*12):(i+1)*12]['DATE'].dt.month, unrate[(i*12):(i+1)*12]['UNRATE'], label=str(1948+i))
plt.show()
```

## Add more curves


- Set the graph display area to 10 inches wide and 6 inches high.
